# VarBranch ablation experiment

Controls the branch variable selection strategy. The default -1 setting makes an automatic choice, depending on problem characteristics. Available alternatives are Pseudo Reduced Cost Branching (0), Pseudo Shadow Price Branching (1), Maximum Infeasibility Branching (2), and Strong Branching (3).

```julia
"Solve the optimization problem for n and k - pass a vector of parameters to the solver, return node count and time"
function odd_x_nodes(n::Int, k::Int; parameters::Vector=[])
    #model = Model(Gurobi.Optimizer)
    model = read_from_file("markshare2.mps")
    set_optimizer(model, Gurobi.Optimizer)
    set_optimizer_attributes(model, parameters...)
    # Define optimization problem
    #@variable(model, x[1:(n+1)], Bin)
    #@objective(model, Min, x[n+1])
    #@constraint(model, 2 * sum(x[i] for i = 1:n) + x[n + 1] == 2 * k + 1)
    # Solve the problem
    solvetime = @elapsed optimize!(model)
    # Extract the number of nodes visited
    nodes = MOI.get(model, MOI.NodeCount())
    return nodes, solvetime
end
```

```julia
result = [odd_x_nodes(20, 10, parameters = ["OutputFlag" => 0, "TimeLimit" => 180, "Cuts" => i]) for i = [0, 1, 2, 3]]
```

```julia
bar([0, 1, 2, 3], log10.(first.(result)), yticks=([1, 2, 3, 4, 5, 6, 7], ["1e$i" for i = 1:7]),
    xticks = ([0, 1, 2, 3], ["No cuts", "Moderate", "Aggressive", "Very aggressive"]), label="", ylabel="Nodes visited")
```

```julia
bar([0, 1, 2, 3], (last.(result)), fill="orange", linecolor="orange",
    xticks = ([0, 1, 2, 3], ["Cuts = $j" for j = 0:3]), label="", ylabel="Solve time (s)")
```
